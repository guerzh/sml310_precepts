```python
import torch
from torch.autograd import Variable
import numpy as np

def f(x):
    return torch.pow((2*x - 5), 2) + torch.pow((3*x - 10), 2)
    
    
x = Variable(torch.from_numpy(np.array([0.0])), requires_grad = True)
```

We can now compute the gradient automatically:

```python
y = f(x)
y.backward()
x.grad.data
```

Now, we can run gradient descent:

```python
for i in range(1000):
    y = f(x)
    y.backward()
    x.data = x.data - 0.01*x.grad.data 
    x.grad.data.zero_()  # Need to zero out x.grad.data, else 
                         # the values would accumulate
    print(x.data.numpy()[0], y.data.numpy()[0])
```
